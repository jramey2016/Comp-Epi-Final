```python
import pandas as pd
from IPython.display import display
import matplotlib.pyplot as plt
import numpy as np
```

# US 2015 - 2019 Influenza Cases

```python
# Load the data from the Excel file
df_2015_to_2019 = pd.read_excel('US_2015-2019_influenza_Data.xlsx', engine='openpyxl')

# Group the data by year and ISO_WEEK, calculate the sum (sentinel and non-sentinel) of INF_ALL
df_agg9 = df_2015_to_2019.groupby(['ISO_YEAR', 'ISO_SDATE']).agg({'INF_ALL': 'sum', 'INF_A': 'sum', 'INF_B': 'sum'}).reset_index()

# Extract only the date portion 
df_agg9['ISO_SDATE'] = df_agg9['ISO_SDATE'].str.split('T').str[0]

# Create a plot
plt.figure(figsize=(16, 6))
plt.plot(df_agg9['ISO_SDATE'], df_agg9['INF_ALL'], label='Cumulative Cases')
plt.plot(df_agg9['ISO_SDATE'], df_agg9['INF_A'], label='Influenza A Cases')
plt.plot(df_agg9['ISO_SDATE'], df_agg9['INF_B'], label='Influenza B Cases')

plt.xlabel('DATE')
plt.ylabel('Number of Cases')
plt.title('US - Influenza Cases from 2015 to 2019')

xtick_pos = np.arange(0, len(df_agg9['ISO_SDATE']), 20)
plt.xticks(xtick_pos, df_agg9['ISO_SDATE'][xtick_pos], rotation=45)
plt.legend()
plt.grid()
plt.show()
```

```python
# Load the data from the Excel file
df_2020_to_2022 = pd.read_excel('US_2020-2022_influenza_Data.xlsx', engine='openpyxl')

# Group the data by year and ISO_WEEK, calculate the sum (sentinel and non-sentinel) of INF_ALL for each group
df_agg11 = df_2020_to_2022.groupby(['ISO_YEAR', 'ISO_SDATE']).agg({'INF_ALL': 'sum', 'INF_A': 'sum', 'INF_B': 'sum'}).reset_index()

# Extract only the date portion 
df_agg11['ISO_SDATE'] = pd.to_datetime(df_agg11['ISO_SDATE'].str.split('T').str[0])

# Create a plot using 
plt.figure(figsize=(16, 6))

plt.plot(df_agg11['ISO_SDATE'], df_agg11['INF_ALL'], label='Cumulative Cases')
plt.plot(df_agg11['ISO_SDATE'], df_agg11['INF_A'], label='Influenza A Cases')
plt.plot(df_agg11['ISO_SDATE'], df_agg11['INF_B'], label='Influenza B Cases')

#important dates
plt.axvline(pd.Timestamp('2020-03-13'), color='black', alpha=0.4)
plt.text(pd.Timestamp('2020-03-13'), 45000, 'March 13, 2020 - \nThe United States declares a national emergency due \nto the COVID-19 pandemic.', rotation=0, color='black', fontweight='bold')

plt.axvline(pd.Timestamp('2020-04-03'), color='black', alpha=0.4)
plt.text(pd.Timestamp('2020-04-03'), 38000, 'April 3, 2020 - The Centers for Disease Control and Prevention (CDC)\n recommends that people wear cloth face coverings in public settings.', rotation=0, color='black', fontweight='bold')

plt.axvline(pd.Timestamp('2020-05-01'), color='black', alpha=0.4)
plt.text(pd.Timestamp('2020-05-01'), 30000, 'May-June 2020 - States begin phased reopening plans, \ngradually lifting restrictions on businesses and social activities. \nThe timeline and extent of reopening vary across states.', rotation=0, color='black', fontweight='bold')

plt.axvline(pd.Timestamp('2020-11-01'), color='black', alpha=0.4)
plt.text(pd.Timestamp('2020-11-01'), 23000, 'Fall 2020 - As the country enters the flu season, some states reintroduce \nrestrictions to prevent a surge in COVID-19 cases, including limiting indoor gatherings\n and requiring masks in public settings.', rotation=0, color='black', fontweight='bold')

plt.xlabel('DATE')
plt.ylabel('Number of Cases')
plt.title('US - Influenza Cases from 2020 to 2022')

plt.xlim(df_agg11['ISO_SDATE'].min(), df_agg11['ISO_SDATE'].max())

xtick_pos = np.arange(0, len(df_agg11['ISO_SDATE']), 1)
plt.xticks(xtick_pos, df_agg11['ISO_SDATE'][xtick_pos], rotation=45)

# Clip x-axis to the first and last data points
plt.gca().set_xlim(left=df_agg11['ISO_SDATE'].iloc[0], right=df_agg11['ISO_SDATE'].iloc[-1])

plt.show()
```

# US 2020 - 2022 influenza cases with specimine collection

```python
# Load the data from the Excel file
df_2020_to_2022 = pd.read_excel('US_2020-2022_influenza_Data.xlsx', engine='openpyxl')

# # Group the data 
df_agg11 = df_2020_to_2022.groupby(['ISO_YEAR', 'ISO_SDATE']).agg({'INF_ALL': 'mean', 'INF_A': 'mean', 'INF_B': 'mean'}).reset_index()

df_agg_samples = df_2020_to_2022.groupby(['ISO_YEAR','ISO_SDATE']).agg({'SPEC_PROCESSED_NB' :'sum'}).reset_index()
# # Extract only the date portion 
df_agg11['ISO_SDATE'] = df_agg11['ISO_SDATE'].str.split('T').str[0]
df_agg_samples['ISO_SDATE'] = df_agg_samples['ISO_SDATE'].str.split('T').str[0]
# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(16, 16))

# Plot the line graphs in the first subplot
ax1.plot(df_agg11['ISO_SDATE'], df_agg11['INF_ALL'], label='Cumulative Cases')
ax1.plot(df_agg11['ISO_SDATE'], df_agg11['INF_A'], label='Influenza A Cases')
ax1.plot(df_agg11['ISO_SDATE'], df_agg11['INF_B'], label='Influenza B Cases')
ax1.set_xlabel('DATE')
ax1.set_ylabel('Number of Cases')
ax1.set_title('US - Influenza Cases from 2020 to 2022')
xtick_pos = np.arange(0, len(df_agg11['ISO_SDATE']), 10)
ax1.set_xticks(xtick_pos)
ax1.set_xticklabels(df_agg11['ISO_SDATE'][xtick_pos], rotation=45)
ax1.legend()
ax1.grid()

# Plot the bar graph in the second subplot
ax2.bar(df_agg_samples['ISO_SDATE'], df_agg_samples['SPEC_PROCESSED_NB'], label='Samples Processed')
ax2.set_xlabel('DATE')
ax2.set_ylabel('Number of Samples')
ax2.set_title('US - Samples Processed from 2020 to 2022')
xtick_pos = np.arange(0, len(df_agg_samples['ISO_SDATE']), 10)
ax2.set_xticks(xtick_pos)
ax2.set_xticklabels(df_agg_samples['ISO_SDATE'][xtick_pos], rotation=45)
ax2.legend()
ax2.grid()

# Adjust the spacing between the subplots
fig.subplots_adjust(hspace=0.4)

# Show the combined graph with both line and bar plots
plt.show()
```
